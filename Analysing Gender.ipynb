```python
import pandas as pd
```

```python
df = pd.read_csv('BlackFriday.csv')

del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
for column in df.columns:
    print(df[column].nunique(), "\t: ", column)
```

```python
data = pd.DataFrame({'Ratio' : [len(df[df['Gender'] == 'M']), len(df[df['Gender'] == 'F'])]}, index = ['Male', 'Female'])

data.plot.pie(y = 'Ratio', figsize = (6,6), autopct = "%.1f", colors = ['lightblue', 'pink'])
```

```python
df.groupby('Gender').size().plot(kind = 'pie', 
                                 autopct = "%.1f",
                                 title = 'Gender Ratio',
                                 figsize = (6,6))
```

```python
df.groupby('Gender').size().plot(kind = 'bar', figsize = (6,6), color = ['pink','lightblue'])
```

```python
df.groupby('Gender').size()
```

```python
df.groupby('Gender').sum()['Purchase'].plot(kind = 'pie', autopct = "%0.1f")
```

```python
df.groupby('Gender')['Purchase'].mean().plot(kind = 'pie', autopct = "%0.1f")
```
